### <b><font color='#56b4e9'> Detailed Column Descriptions </font></b>
This section offers a comprehensive explanation of each data column within the dataset, focusing on predictive scores, structural metrics, and functional annotations:

- **`esm_1b_masked_marginals`**: Predicted score calculated using a masked marginal scoring approach applied to the ESM-1b model (https://doi.org/10.1073/pnas.2016239118).

- **`esm_if_masked_marginals`**: Predicted score calculated using a masked marginal scoring approach applied to the ESM-IF model (https://proceedings.mlr.press/v162/hsu22a.html).

- **`exposure_plddt`**: pLDDT (predicted local distance difference test) scores for each residue of the AlphaFold model extracted from the AlphaFold DB (https://doi.org/10.1093/nar/gkab1061). This score is in the range of [0, 100] and represents the confidence of the predicted structure.

- **`exposure_SS`**: DSSP secondary structure (SS) classes from the DSSP software (https://doi.org/10.1002/bip.360221211). H is 'Alpha helix (4-12)', B is 'Isolated beta-bridge residue', E is 'Strand', G is '3-10 helix', I is 'Pi helix', T is 'Turn', S is 'Bend', - is 'None'.

- **`exposure_ASA`**: Accessible surface area (ASA) or solvent-accessible surface area (SASA) values, measuring the surface area of residues exposed to solvent. Calculated using DSSP (https://doi.org/10.1002/bip.360221211).

- **`exposure_rASA`**: Relative Accessible Surface Area (rASA) normalized to the maximum possible exposure (https://doi.org/10.1002/prot.340200303).

- **`exposure_CN`**: Contact number (CN), quantifying residue exposure as the number of neighboring C-alpha atoms around its C-alpha atom within a radius of 12.0 Angstrom from C-alpha.

- **`exposure_HSE_U` and `exposure_HSE_D`**: Half-Sphere Exposure (Up or Down), the exposure of residues on the upper or lower half-sphere relative to a defined plane on the C-beta with a radius of 12.0 Angstrom (https://doi.org/10.1002/prot.20379).

- **`exposure_Vol_U_CB` and `exposure_Vol_D_CB`**: Visible Volume measure (Up or Down) within a radius of 12.0 Angstrom from the C-beta atom and a disk radius of 3.8 Angstrom. (https://doi.org/10.1002/prot.24277).

- **`exposure_Exp_U_CB` and `exposure_Exp_D_CB`**: Solvent Exposure measure (Up or Down) within a radius of 12.0 Angstrom from the C-beta atom and a disk radius of 3.8 Angstrom. (https://doi.org/10.1002/prot.24277).

- **`exposure_WCN_CA` and `exposure_WCN_SC`**: Weighted Contact Number (WCN) calculated on the C-alpha atom or side chain, evaluating the density of neighboring atoms around the C-alpha atom or the side chain atoms of the residue, respectively.

- **`spot_disorder`**: Binary indication of a residue being disordered, based on its sequence context and structural features. Data lifted from Alderson et al. 2023 ( https://doi.org/10.1073/pnas.2304302120).

- **`dsbdetector_cys_in_disulfide`**: Binary indication of cysteine residues being involved in disulfide bonds based on the analysis of their spatial proximity and angle alignment.

- **`func_esms_residue_class`**: Functional classification of residues based on Func-ESMs threshold method, categorizing their role within the protein. 0 is 'tolerant', 1 is 'functionally relevant', 2 is 'structurally critical', and 4 is 'mixed'. As described in Cagiada, Jonsson and Lindorff-Larsen.

- **`func_esms_variant_class`**: Functional classification of variants based on Func-ESMs threshold method, categorizing their role within the protein. 0 is 'WT-like', 1 is 'stable-but-inactive', and 2 is 'total-loss'. As described in Cagiada, Jonsson and Lindorff-Larsen.

- **`clinvar_clinical_significance`**: Character field summarizing the clinical assessment of a variant in a simplified form parsed from ClinVar. As described in Cagiada, Jonsson and Lindorff-Larsen.

This detailed column information enhances understanding and enables more effective utilization of the dataset for comprehensive protein analysis.

```python
#@title <b><font color='#f0e422'>PRELIMINARY OPERATIONS:</font></b> Load pipeline functions
#@markdown **Run the cell to load the required functions.**

import pandas as pd
import urllib.request
import json
import os
import re
import requests
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
from IPython.display import display, Markdown

## dictionaries
alphabetAA_L_D={'-':0,'_' :0,'A':1,'C':2,'D':3,'E':4,'F':5,'G':6,'H':7,'I':8,'K':9,'L':10,'M':11,'N':12,'P':13,'Q':14,'R':15,'S':16,'T':17,'V':18,'W':19,'Y':20}
alphabetAA_D_L={v: k for k, v in alphabetAA_L_D.items()}

AA_to_hydrophobicity_scores={'A':44,'C':50,'D':-37,'E':-12,'F':96,'G':0,'H':-16,'I':100,'K':-30,'L':99,'M':74,'N':-35,'P':-46,'Q':-14,'R':-20,'S':-6,'T':13,'V':78,'W':90,'Y':57}

alphabet_class_variant_d_l={0:'WT-like',1:'SBI',2:'Total-loss'}
alphabet_class_residue_d_l={0:'WT-like',1:'Functional site',2:'Total-loss',5:'Not assgined'}

aa_order_alphabetical = pd.Series(["A", "C", "D", "E", "F", "G", "H", "I", "K", "L", "M",
           "N", "P", "Q", "R", "S", "T", "V", "W", "Y"])

alphabetAA_L_D_reordered={0:'R',1:'H',2:'K',3:'D',4:'E',5:'S',6:'T',7:'N',8:'Q',9:'C',10:'G',11:'P',12:'A',13:'V',14:'I',15:'L',16:'M',17:'F',18:'Y',19:'W'}
alphabetAA_D_D_reordered={14:0,6:1,8:2,2:3,3:4,15:5,16:6,11:7,13:8,1:9,5:10,12:11,0:12,17:13,7:14,9:15,10:16,4:17,19:18,18:19}

## Functions
def fetchUniProtData(uniprot, option=None):
    """
    Fetches data from the UniProtKB database using the REST API selectively based on the provided option.
    If the gene name contains spaces, they are replaced with hyphens. It also retrieves 'alternativeNames'
    and 'submissionNames' from 'proteinDescription' if they exist, extracting all 'value' fields.
    """
    urlTemplateUniProt = f"https://rest.uniprot.org/uniprotkb/{uniprot}.json"

    try:
        with urllib.request.urlopen(urlTemplateUniProt) as link:
            dataUniProt = json.loads(link.read().decode())
    except Exception as e:
        raise Exception(f"An error occurred accessing the UniProtKB REST API: {e}")

    def get_gene_name(data):
        gene_name = next((gene.get('geneName', {}).get('value', 'Unknown')
                          for gene in data.get('genes', [])), 'Unknown')
        return gene_name.replace(' ', '-')

    def get_names_list(data, key):
        # Extracts a list of names from the given key in proteinDescription
        names_data = data.get('proteinDescription', {}).get(key, [])
        names_list = [name.get('fullName', {}).get('value') for name in names_data if 'fullName' in name and 'value' in name['fullName']]

        # If there's only one name, return it directly, else return the list of names or 'Unknown'
        if len(names_list) == 1:
            return names_list[0]
        else:
            return names_list if names_list else 'Unknown'

    fetchedData = {
        'primaryAccession': dataUniProt.get('primaryAccession', 'Unknown'),
        'uniProtkbId': dataUniProt.get('uniProtkbId', 'Unknown'),
        'sequenceVersion': dataUniProt.get('entryAudit', {}).get('sequenceVersion', 'Unknown'),
        'geneName': get_gene_name(dataUniProt),
        'recommendedName': dataUniProt.get('proteinDescription', {}).get('recommendedName', {}).get('fullName', {}).get('value', 'Unknown'),
        'alternativeNames': get_names_list(dataUniProt, 'alternativeNames'),
        'submissionNames': get_names_list(dataUniProt, 'submissionNames'),
        'organismScientificName': dataUniProt.get('organism', {}).get('scientificName', 'Unknown'),
        'organismCommonName': dataUniProt.get('organism', {}).get('commonName', 'Unknown'),
        'sequence': dataUniProt.get('sequence', {}).get('value', 'Unknown')
    }

    # Replace unwanted symbols in 'recommendedName', 'alternativeNames', and 'submissionNames'
    for key in ['recommendedName', 'alternativeNames', 'submissionNames']:
        if isinstance(fetchedData[key], str):
            fetchedData[key] = fetchedData[key].replace(';', '').replace(':', '').replace('[', '').replace(']', '')

    optionsMap = {
        'primaryAccession': 'primaryAccession',
        'uniProtkbId': 'uniProtkbId',
        'sequenceVersion': ('entryAudit', 'sequenceVersion'),
        'geneName': get_gene_name,
        'recommendedName': ('proteinDescription', 'recommendedName', 'fullName', 'value'),
        'alternativeNames': lambda data: get_names_list(data, 'alternativeNames'),
        'submissionNames': lambda data: get_names_list(data, 'submissionNames'),
        'organismScientificName': ('organism', 'scientificName'),
        'organismCommonName': ('organism', 'commonName'),
        'sequence': ('sequence', 'value')
    }

    if option is None:
        return fetchedData
    else:
        if option in optionsMap:
            if callable(optionsMap[option]):
                return optionsMap[option](dataUniProt)
            else:
                keys = optionsMap[option]
                if isinstance(keys, tuple):
                    value = dataUniProt
                    for key in keys:
                        value = value.get(key, 'Unknown')
                        if value == 'Unknown':
                            break
                    return value
                else:
                    return dataUniProt.get(keys, 'Unknown')
        else:
            raise ValueError(f"Invalid option provided: {option}")

def make_default_outfolder(uniprot_id, base_prism_dir='./output/'):
    """
    Creates a directory path for output files based on a UniProt ID.

    This function constructs a directory path by appending segments of the UniProt ID to the base directory path.
    If the resulting directory does not exist, it is created.

    Parameters:
    uniprot_id (str): The UniProt ID to base the directory path on.
    base_prism_dir (str): The base directory path. Defaults to './output/' in the current directory.

    Returns:
    str: The path of the output directory.
    """
    if uniprot_id != 'None':
        outfolder = os.path.join(base_prism_dir, uniprot_id)
    else:
        outfolder = os.path.join(base_prism_dir, 'NA')

    if not os.path.exists(outfolder):
        os.makedirs(outfolder)

    return outfolder

def download_erda_file(sharelink, uniprot_id, gene_name):
    """
    Download a specific ERDA file based on the sharelink, uniprot ID, and gene name.

    Parameters:
    sharelink (str): The base sharelink URL from ERDA.
    uniprot_id (str): The uniprot ID in the form A0A023T787.
    gene_name (str): The gene name associated with the uniprot ID.

    Returns:
    str: The path of the downloaded file or None if the download failed.
    """
    # Partition the uniprot ID
    partitioned_uniprot = f"{uniprot_id[:2]}/{uniprot_id[2:4]}/{uniprot_id[4:6]}"

    # Generate the new directory structure
    new_directory = f"erda/{partitioned_uniprot}"

    # Construct the file name
    file_name = f"prism_func_{uniprot_id}-{gene_name}.txt"

    # Construct the final URL
    file_url = f"https://sid.erda.dk/share_redirect/{sharelink}/{new_directory}/{file_name}"

    # Determine the output folder
    output_folder = make_default_outfolder(uniprot_id)
    local_filename = os.path.join(output_folder, file_name)

    # Download the file
    response = requests.get(file_url, stream=True)
    if response.status_code != 200:
        print(f"Failed to download file: {file_url}")
        return None

    with open(local_filename, 'wb') as f:
        for chunk in response.iter_content(chunk_size=8192):
            f.write(chunk)

    return local_filename

def load_data_variant(data: str, wt_seq: str, start_gap: int = 0, score_load: str = '') -> np.ndarray:
    """
    Load mutation data from a txt file, align it with the wild-type sequence, and return a score matrix.

    Args:
        data (str): Path to the txt file.
        wt_seq (str): Wild-type sequence.
        start_gap (int, optional): Gap value to start from (default is 0).
        score_load (str, optional): Specify the type of score to load ("esm", "rasp", "gemme", or "func").

    Returns:
        np.ndarray: 2D array of scores aligned to the wild-type sequence.
    """

    # Read the data file
    df = pd.read_csv(data, comment='#', sep=r'\s+')

    # Extract the reference amino acid (aa_ref)
    df['aa_ref'] = df['variant'].str[0]

    # Extract the residue number (resi)
    df['resi'] = df['variant'].str[1:-1].astype(int)

    # Extract the variant amino acid (aa_var)
    df['aa_var'] = df['variant'].str[-1]

    df = df[df["aa_var"] != "*"]

    # Load mutation data
    mutation_load = np.array(df['variant'])

    # Identify the score column based on score_load
    score_column_mapping = {
        'esm-1b': 'esm_1b_masked_marginals',
        'esm-if': 'esm_if_masked_marginals',
        'func-var': 'func_esms_variant_class',
    }

    if score_load not in score_column_mapping:
        raise ValueError(f"Invalid score load value. Expected one of {list(score_column_mapping.keys())}")

    score_column = df[score_column_mapping[score_load]].values

    # Initialize the score matrix
    scores = np.full((len(wt_seq), 20), np.nan)

    # Fill the scores matrix
    for mutation, score in zip(mutation_load, score_column):
        if mutation[-1] != '=':
            pos = int(mutation[1:-1]) - 1 + start_gap
            aa_index = alphabetAA_L_D[mutation[-1]] - 1
            scores[pos, aa_index] = float(score)

    return scores

def load_data_residue(data: str, wt_seq: str, start_gap: int = 0) -> np.ndarray:
    """
    Load residue data from a txt file, align it with the wild-type sequence, and return a class array.

    Args:
        data (str): Path to the txt file.
        wt_seq (str): Wild-type sequence.
        start_gap (int, optional): Gap value to start from (default is 0).

    Returns:
        np.ndarray: 1D array of residue classes aligned to the wild-type sequence.
    """

    # Read the data file
    df = pd.read_csv(data, comment='#', sep=r'\s+')

    # Extract the reference amino acid (aa_ref)
    df['aa_ref'] = df['variant'].str[0]

    # Extract the residue number (resi)
    df['resi'] = df['variant'].str[1:-1].astype(int)

    # Extract the variant amino acid (aa_var)
    df['aa_var'] = df['variant'].str[-1]

    df = df[df["aa_var"] != "*"]

    # Remove duplicates, keeping the first occurrence for each residue
    df = df.drop_duplicates(subset=['resi'])

    # Initialize the residue class array
    residue_classes = np.full(len(wt_seq), np.nan)

    # Fill the residue class array
    for idx, row in df.iterrows():
        pos = row['resi'] - 1 + start_gap
        if pos < len(residue_classes):
            residue_classes[pos] = row['func_esms_residue_class']

    return residue_classes

def remove_WT_score(score: np.ndarray, WT_seq: str) -> np.ndarray:
    """
    Remove wild-type (WT) scores from a 2D score array.

    Args:
        score (np.ndarray): 2D array of scores.
        WT_seq (str): Wild-type sequence.

    Returns:
        np.ndarray: Updated 2D array with WT scores set to NaN.
    """
    for i in range(len(WT_seq)):
        score[i, alphabetAA_L_D[WT_seq[i]] - 1] = np.nan
    return score

def heatmap(ext_score: np.ndarray, label_cm: str, WT_mask: np.ndarray, lim_down: float, lim_up: float,
            color: str, x: tuple, uniprot_id: str, title_figure: str = '',
            WT_cmap: str = 'gnuplot_r', nan_cmap: str = 'Dark2_r', figs: list = [30, 7],
            xtick_spacing: int = 10, font_size: int = 14, colorbar_padding: float = 0.15):
    """
    Create and display a heatmap of scores with masked and WT values.

    Args:
        ext_score (numpy.ndarray): The input scores array.
        label_cm (str): Label for the colorbar.
        WT_mask (numpy.ndarray): Mask for WT values.
        lim_down (float): Lower limit for color scale.
        lim_up (float): Upper limit for color scale.
        color (str): Colormap for the heatmap.
        x (tuple): Tuple representing x-axis limits (start, end).
        uniprot_id (str): UniProt ID for naming and saving files.
        title_figure (str, optional): Title of the figure. Default is ''.
        WT_cmap (str, optional): Color for WT values. Defaults to 'coolwarm'.
        nan_cmap (str, optional): Colormap for NaN values. Defaults to 'Dark2_r'.
        figs (list, optional): Figure size. Defaults to [30, 7].
        xtick_spacing (int, optional): Spacing between x-axis ticks. Defaults to 10.
        font_size (int, optional): Font size for labels. Defaults to 14.
        colorbar_padding (float, optional): Padding for the colorbar. Defaults to 0.125.

    Returns:
        None
    """
    where_save = make_default_outfolder(uniprot_id)

    # Set Matplotlib font and label sizes
    mpl.rcParams['font.size'] = font_size
    mpl.rcParams['xtick.labelsize'] = font_size
    mpl.rcParams['ytick.labelsize'] = font_size
    mpl.rcParams['axes.labelsize'] = font_size

    # Create a copy of ext_score as score
    score = np.copy(ext_score)

    # Reorder columns in score using alphabetAA_D_D_reordered
    for i in range(score.shape[1]):
        score[:, alphabetAA_D_D_reordered[i]] = ext_score[:, i]

    # Create a copy of ext_score as score_nan filled with infinity
    score_nan = np.copy(ext_score)
    score_nan[:] = np.inf

    # Fill score_nan with 1 where ext_score contains NaN and the corresponding WT_mask condition is met
    for i in range(score.shape[0]):
        for j in range(score.shape[1]):
            if np.isnan(ext_score[i, j]) and WT_mask[i] != alphabetAA_D_L[j + 1]:
                score_nan[i, alphabetAA_D_D_reordered[j]] = 1

    # Modify the colormap for WT scores (dark gray)
    cmap = plt.get_cmap(nan_cmap)
    cmap.set_bad(color='gray')

    # Create a figure and axes
    fig, ax = plt.subplots(figsize=(figs[0], figs[1]))

    # Create ylabels
    ylabels = [alphabetAA_L_D_reordered[i] for i in range(0, 20)]
    ax.set_yticks([i for i in range(0, 20)])
    ax.set_yticklabels(ylabels, fontsize=font_size)

    # Create and display the heatmap with the modified colormap for WT scores
    pos = ax.imshow(np.transpose(score), aspect='auto', cmap=f'{colour_map}_r', vmin=lim_down, vmax=lim_up, interpolation='nearest')

    # Create a divider for the colorbar axes
    divider = make_axes_locatable(ax)

    # Append a new axes for the colorbar with specified padding
    cax = divider.append_axes("right", size="1.5%", pad=colorbar_padding)

    # Add a colorbar with the specified label
    cbar = plt.colorbar(pos, cax=cax, label=label_cm)
    cbar.ax.tick_params(labelsize=font_size)  # Set colorbar label font size

    # Configure plot appearance
    ax.set_xlabel("residue", fontsize=font_size)
    ax.set_title(title_figure, fontsize=font_size)

    start, end = ax.get_xlim()
    ax.xaxis.set_ticks(np.arange(start + 0.5, end, xtick_spacing))
    ax.tick_params(axis='x', rotation=90)
    ax.tick_params(axis='y', rotation=90)

    # Create a temporary array filled with infinity and fill it based on WT_mask and alphabetAA_D_D_reordered
    tmp = np.empty((len(WT_mask), 20), dtype=float)
    tmp[:] = np.inf

    for i in range(len(WT_mask)):
        tmp[i, alphabetAA_D_D_reordered[alphabetAA_L_D[WT_mask[i]] - 1]] = 1

    # Overlay WT and NaN masks on the heatmap
    ax.imshow(np.transpose(score_nan), cmap=nan_cmap, aspect='auto', interpolation='nearest')
    ax.imshow(np.transpose(tmp), cmap=WT_cmap, aspect='auto', interpolation='nearest')

    # Save the figure
    #plt.savefig(os.path.join(where_save, uniprot_id + '_' + 'heatmap.png'), dpi=300, bbox_inches='tight')

    # Display the figure
    plt.show()

def functional_var_heatmap(ext_score: np.ndarray, WT_mask: np.ndarray, x: tuple, uniprot_id: str, title_figure: str = '',
                           WT_cmap: str = 'gnuplot_r', nan_cmap: str = 'Dark2_r', figs: list = [30, 7],
                           xtick_spacing: int = 10, font_size: int = 14) -> None:
    """
    Create and save a heatmap visualizing the extent score with specific color mappings for WT and NaN values.

    Args:
        ext_score (numpy.ndarray): The input scores array.
        WT_mask (numpy.ndarray): Mask for WT (Wild Type) values.
        x (tuple): Tuple representing x-axis limits (start, end).
        uniprot_id (str): UniProt ID for naming and saving files.
        WT_cmap (str, optional): Colormap for WT values. Defaults to 'gnuplot_r'.
        nan_cmap (str, optional): Colormap for NaN values. Defaults to 'Dark2_r'.
        figs (list, optional): Figure size as a list of two values [width, height]. Defaults to [30, 7].
        xtick_spacing (int, optional): Spacing between x-axis ticks. Defaults to 10.
        font_size (int, optional): Font size for labels and ticks. Defaults to 14.

    Returns:
        None: The function saves the heatmap as an image file and also displays it.
    """
    where_save = make_default_outfolder(uniprot_id)

    # Set Matplotlib font and label sizes
    mpl.rcParams['font.size'] = font_size
    mpl.rcParams['xtick.labelsize'] = font_size
    mpl.rcParams['ytick.labelsize'] = font_size
    mpl.rcParams['axes.labelsize'] = font_size

    # Prepare the score array by reordering columns
    score = np.copy(ext_score)

    # Reorder columns in score using alphabetAA_D_D_reordered
    for i in range(score.shape[1]):
        score[:, alphabetAA_D_D_reordered[i]] = ext_score[:, i]

    # Create a copy of ext_score as score_nan filled with infinity
    score_nan = np.copy(ext_score)
    score_nan[:] = np.inf

    # Fill score_nan with 1 where ext_score contains NaN and the corresponding WT_mask condition is met
    for i in range(score.shape[0]):
        for j in range(score.shape[1]):
            if np.isnan(ext_score[i, j]) and WT_mask[i] != alphabetAA_D_L[j + 1]:
                score_nan[i, alphabetAA_D_D_reordered[j]] = 1

    # Modify the colormap for WT scores (dark gray)
    cmap = plt.get_cmap(nan_cmap)
    cmap.set_bad(color='gray')

    # Create a figure and axes
    fig, ax = plt.subplots(figsize=(figs[0], figs[1]))

    # Set y-axis labels
    ylabels = [alphabetAA_L_D_reordered[i] for i in range(0, 20)]
    ax.set_yticks([i for i in range(0, 20)])
    ax.set_yticklabels(ylabels, fontsize=font_size)

    # Determine the number of unique colors required
    num_colors = len(np.unique(score)) - 1 if np.isnan(np.unique(score)).any() else len(np.unique(score))

    # Display the heatmap with appropriate color map
    if num_colors==3:
      pos=ax.imshow(np.transpose(score), aspect = 'auto', cmap=mpl.colors.ListedColormap(["#009e74", "#56b4e9","#d55c00"]),interpolation='nearest')
    elif num_colors==4:
      pos=ax.imshow(np.transpose(score), aspect = 'auto', cmap=mpl.colors.ListedColormap(["#009e74", "#56b4e9","#d55c00","#f0e442"]),interpolation='nearest')

    # Set colormap for bad (NaN) values
    current_cmap = plt.get_cmap()
    current_cmap.set_bad(color='gray')

    # Configure plot appearance
    ax.set_xlabel("residue", fontsize=font_size)
    ax.set_title(title_figure, fontsize=font_size)

    start, end = ax.get_xlim()
    ax.xaxis.set_ticks(np.arange(start + 0.5, end, xtick_spacing))
    ax.tick_params(axis='x', rotation=90)
    ax.tick_params(axis='y', rotation=90)

    # Create a temporary array filled with infinity and fill it based on WT_mask and alphabetAA_D_D_reordered
    tmp = np.empty((len(WT_mask), 20), dtype=float)
    tmp[:] = np.inf

    for i in range(len(WT_mask)):
        tmp[i, alphabetAA_D_D_reordered[alphabetAA_L_D[WT_mask[i]] - 1]] = 1

    # Overlay WT and NaN masks on the heatmap
    ax.imshow(np.transpose(score_nan), cmap=nan_cmap, aspect='auto', interpolation='nearest')
    ax.imshow(np.transpose(tmp), cmap=WT_cmap, aspect='auto', interpolation='nearest')

    # Save the figure
    #plt.savefig(os.path.join(where_save, uniprot_id + '_variant_map.png'), dpi=400, bbox_inches='tight')

    # Display the figure
    plt.show()

def functional_res_heatmap(ext_score: np.ndarray, x: tuple, uniprot_id: str, title_figure: str = '',
                           figs: list = [20, 1], xtick_spacing: int = 10, font_size: int = 14) -> None:
    """
    Create and save a heatmap of reordered residue classes.

    Args:
        ext_score (np.ndarray): Array of residue classes.
        x (tuple): Tuple representing x-axis limits (start, end).
        uniprot_id (str): UniProt ID for naming and saving files.
        title_figure (str, optional): Title of the figure. Default is ''.
        figs (list, optional): Figure size as a list of two values [width, height]. Default is [20, 1].
        xtick_spacing (int, optional): Spacing between x-axis ticks. Default is 10.
        font_size (int, optional): Font size for labels and ticks. Defaults to 14.

    Returns:
        None: The function saves the figure as an image file and also displays it.
    """
    where_save = make_default_outfolder(uniprot_id)

    # Set Matplotlib font and label sizes
    mpl.rcParams['font.size'] = font_size
    mpl.rcParams['xtick.labelsize'] = font_size
    mpl.rcParams['ytick.labelsize'] = font_size
    mpl.rcParams['axes.labelsize'] = font_size

    # Create a figure and axis
    fig, ax = plt.subplots(1, 1, figsize=(figs[0], figs[1]))
    fig.tight_layout()  # Adjust layout

    # Determine the colormap based on the unique values in ext_score
    if 4. in ext_score:
      cmap = mpl.colors.ListedColormap(["#009e74", "#56b4e9","#d55c00","#5A5A5A","#5A5A5A"])
    else:
      cmap = mpl.colors.ListedColormap(["#009e74", "#56b4e9", "#d55c00"])

    # Display the heatmap
    ax.imshow(ext_score.reshape(1, -1), aspect='auto', cmap=cmap)

    # Set y-axis labels
    ax.set_ylabel('FunC-ESMs', fontsize=font_size)

    # Remove y-axis ticks
    ax.set_yticks([])

    # Set x-axis limits
    plt.xlim(x[0] - 0.5, x[1] + 0.5)

    # Set x-axis ticks and labels
    start, end = ax.get_xlim()
    ax.xaxis.set_ticks(np.arange(start + 0.5, end, xtick_spacing))
    ax.tick_params(axis='x', rotation=90)

    # Set x-axis label and title
    ax.set_xlabel('residue', fontsize=font_size)
    ax.set_title(title_figure, fontsize=font_size)

    # Save the figure
    #plt.savefig(os.path.join(where_save, uniprot_id + '_residue_map.png'), dpi=400, bbox_inches='tight')

    # Display the figure
    plt.show()
```
